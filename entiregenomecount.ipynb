```python
from collections import Counter
import scipy.stats as stats
import pandas 

"""Open up files and ignore spaces and gene IDs"""

def make_count(file_names):

    counter = {}
    codonlist = []
    
    for f in file_names:
        
        with open(f, 'r') as file:
            seq = ''
            for line in file.readlines():
                if line[0] != '>':
                    seq = seq + line.strip()
                else:
                    seq = seq + ' '
    
    """Count gene sequences"""
    
    for n in range(len(seq)-5):
        codonlist.append(seq[n:n+6])

    for c in codonlist:
        counter[c] = counter.get(c,0) + 1

    return counter
```

```python
genome_file_names = ['entiregenome.txt']
genome_counter = make_count(genome_file_names)
```

```python
codons = list(genome_counter.keys())

for c in codons:
    g_mut_count = genome_counter[c]
    
    print(c, g_mut_count)
```

```python
def do_count(file_names):

    counter = {}
    codonlist = []
    
    for f in file_names:
        
        with open(f, 'r') as file:
            for line in file.readlines(): 
                lr = line.strip()
    
                counter[lr] = counter.get(lr,0) + 1

    return counter
            
```

```python
cancer_file_names = ['delsixnuc.txt', 'inssixnuc.txt']
cancer_counter = do_count(cancer_file_names)
```

```python
codons = list(cancer_counter.keys())

for c in codons:
    c_mut_count = cancer_counter[c]
    
    print(c, c_mut_count)
    
    
```
