```python
import pandas as pd

df = pd.read_csv('martexport1.csv')

# All exons
df
```

```python
# Last exons only

df1 = df.loc[df['Transcript end (bp)']==df['Exon region end (bp)']]

df1
```

```python
# Frameshift mutations 

ef = pd.read_csv('fsposition.csv')

ef
```

```python
# Frameshift mutation pos

def from_ (string):
    return string[string.index('-')+1:]


ef['pos'] = ef.apply(lambda row: from_(row['Mutation genome position']), axis = 'columns')

ef
```

```python
# merge spreadsheets.

merg = pd.merge(ef, df1, left_on='Accession Number', right_on='Transcript stable ID')

merg
```

```python
import numpy as np

def my_test(row):
    return int(row['pos']) >= int(row['Exon region start (bp)'])

merg['outcome'] = merg.apply(lambda row: my_test(row), axis=1)

merg
```
