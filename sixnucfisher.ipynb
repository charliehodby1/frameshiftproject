```python
from collections import Counter
import scipy.stats as stats
import pandas 

"""Open up files and ignore spaces and gene IDs"""

def make_count(file_names):

    counter = {}
    codonlist = []
    
    for f in file_names:
        
        with open('entiregenome.txt', 'r') as file:
            seq = ''
            for line in file.readlines():
                if line[0] != '>':
                    seq = seq + line.strip()
                else:
                    seq = seq + ' '
    
    """Count gene sequences"""
    
    for n in range(len(seq)-5):
        codonlist.append(seq[n:n+6])

    for c in codonlist:
        counter[c] = counter.get(c,0) + 1

    return counter
```

```python
"""List of file names to open"""

cancer_file_names = ['delsixnuc.txt', 'inssixnuc.txt']
cancer_counter = make_count(cancer_file_names)

genome_file_names = ['entiregenome.txt']
genome_counter = make_count(genome_file_names)
```

```python
"""Perform fishers test on counts"""

codons = list(cancer_counter.keys())

for c in codons:
    c_mut_count = cancer_counter[c]
    c_all_count = sum(cancer_counter.values())
    c_other_count = c_all_count - c_mut_count
    print(),
    
    
    g_mut_count = genome_counter[c]
    g_all_count = sum(genome_counter.values())
    g_other_count = g_all_count - g_mut_count
    
    oddsratio, pvalue = stats.fisher_exact([[c_mut_count, c_other_count], [g_mut_count, g_other_count]])
    
    print(c, c_mut_count, c_other_count, g_mut_count, g_other_count, pvalue)

```
